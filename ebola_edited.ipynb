```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "ebola.csv"
```

```python
ebola_data_df = pd.read_csv(csv_file)
ebola_data_df.head(50)
```

```python
ebola_data_df = ebola_data_df.drop(columns=["Country"])
ebola_data_df.head()
```

```python
#ebola_data_df = ebola_data_df.rename(columns={"Indicator":" Total cases"})
ebola_data_df.head()
```

```python
#ebola_data_df.count
ebola_data_df["Indicator"].value_counts()
```

```python
# new_ebola_df = ebola_data_df.replace(
#     {"Cumulative number of confirmed Ebola cases": "Confirmed Ebola cases", "Cumulative number of confirmed, probable and s..": "Confirmed Ebola cases", "Cumulative number of suspected Ebola cases": "Confirmed Ebola cases"})
# new_ebola_df
```

```python
Confimred_cases_df = ebola_data_df.loc[ebola_data_df["Indicator"]=="Cumulative number of confirmed Ebola cases",:].replace(
    {"Cumulative number of confirmed Ebola cases": "Confirmed Ebola cases", "Cumulative number of confirmed, probable and s..": "Confirmed Ebola cases", "Cumulative number of suspected Ebola cases": "Confirmed Ebola cases"})
Confimred_cases_df = Confimred_cases_df.rename(columns={"Indicator":"Confirmed cases"})
Confimred_cases_df.head()
```

```python
Confirmed_death_df = ebola_data_df.loc[ebola_data_df["Indicator"]=="Cumulative number of confirmed Ebola deaths",:].replace({"Cumulative number of confirmed Ebola deaths": "Confirmed Ebola deaths"})
Confirmed_death_df = Confirmed_death_df.rename(columns={"Indicator":"Confirmed deaths"})
Confirmed_death_df.head()
```

```python
new_df = pd.merge(Confimred_cases_df, Confirmed_death_df, on='Date', how='left')
new_df
```

```python
clean_df = new_df.rename(columns={
    "value_x": "confirmed_cases",
    "value_y": "confirmed_deaths",
    "Date": "date"
})
clean_df.head()
```

```python
clean_df1 = clean_df.drop(['Confirmed cases'], axis=1)
clean_df.head()
```

```python
clean_df2 = clean_df1.drop(['Confirmed deaths'], axis=1)
clean_df2
```

```python
confirmed_cases_df = clean_df2.groupby(["date"]).sum()["confirmed_cases"]
confirmed_cases_df.head()
```

```python
confirmed_death_df = clean_df2.groupby(["date"]).sum()["confirmed_deaths"]
```

```python
new_dataframe = pd.DataFrame({"confirmed_cases_df": confirmed_cases_df,
                                "confirmed_death_df": confirmed_death_df
        
                               })
new_dataframe
```

```python
rename_df = new_dataframe.rename(columns={
    "confirmed_cases_df": "total_cases",
    "confirmed_death_df": "total_deaths",
    "date": "date"
    

})
rename_df
```

```python
connection_string = "postgres:1993@localhost:5432/ebola_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
rename_df.to_sql(name="ebola", con=engine, if_exists="append", index=True)
```
